```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 330
mu = 0.4439
sigma = 0.536
T = 1
N = 5896800
dt = T/N
np.random.seed(42)

# Simulate Brownian motion (random steps)
dW = np.random.normal(0,np.sqrt(dt), N)
W = np.cumsum(dW) #Browninan Path

# Simulate stock price path
t = np.linspace(0,T,N)
S = S0 * np.exp((mu - 0.5 *sigma**2 )*t + sigma * W)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, S)
plt.title("Simulated Stock Price (S_t)")
plt.xlabel("Time")
plt.ylabel("S_t")

plt.subplot(1, 2, 2)
plt.plot(t, np.log(S))
plt.title("Log of Stock Price (ln(S_t))")
plt.xlabel("Time")
plt.ylabel("ln(S_t)")

plt.tight_layout()
plt.show()
```

```python
lnS_pred = np.log(S0) + (mu - 0.5 * sigma**2) * t + sigma * W

plt.plot(t, np.log(S), label="Actual ln(S_t)")
plt.plot(t, lnS_pred, label="Itô’s Lemma Approx", linestyle='--')
plt.title("Itô’s Lemma vs Actual ln(S_t)")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Log Price")
plt.grid(True)
plt.show()
```
